```python
# A list holds multiple values in order
vital_signs = ["pulse", "temperature", "SBP", "DBP", "RR", "FiO2", "GCS"]

print("All vitals:", vital_signs)
print("First vital:", vital_signs[0])      # indexing starts at 0
print("Last vital:", vital_signs[-1])      # -1 gets the last item
print("Number of vitals:", len(vital_signs))
```

```python
# A for loop runs a block of code once for each item in a list
for vital in vital_signs:
    print(f"  - {vital}")
```

```python
# Conditionals let code make decisions
def classify_pulse(bpm):
    if bpm < 60:
        return "Bradycardia (too slow)"
    elif bpm > 100:
        return "Tachycardia (too fast)"
    else:
        return "Normal"

# Test it
test_pulses = [45, 78, 130]
for p in test_pulses:
    print(f"  Pulse {p} bpm -> {classify_pulse(p)}")
```

```python
# The three libraries we use most in this programme
import pandas as pd          # data manipulation
import numpy as np           # numerical operations
import matplotlib.pyplot as plt  # plotting

print("Libraries imported successfully!")
```

```python
# ── TUTOR: adjust this path if the file is in Drive ──────────
FILE_PATH = 'EmergencyTriageDataset_Reduced_Dirty.csv'
#FILE_PATH = '/content/drive/MyDrive/ColabNotebooks/CariSurg_Triage_Test/EmergencyTriageDataset_Reduced_Dirty.csv'

df_raw = pd.read_csv(FILE_PATH)
print(f"Dataset loaded: {df_raw.shape[0]} rows x {df_raw.shape[1]} columns")
print(f"Columns: {list(df_raw.columns)}")
```

```python
# Always look at the first few rows immediately after loading
df_raw.head(10)
```

```python
# What data types does pandas think each column is?
print(df_raw.dtypes)
```

```python
# Step 1: What values are actually in the Gender column?
print("Unique values in Gender:")
print(df_raw['Gender'].unique())
print(f"\nTotal unique values: {df_raw['Gender'].nunique()}")
```

```python
# Step 2: Count how many of each value we have
print("Gender value counts:")
print(df_raw['Gender'].value_counts())
```

```python
# Step 3: Build the mapping — every variant maps to 1 or 0
gender_map = {
    'Male':   1,
    'MALE':   1,
    '1':      1,
    'Female': 0,
    'FEMALE': 0,
    '0':      0
}

# Step 4: Apply the mapping to create a new, clean column
# It is best practice to make changes to a new column rather than editing an
# exising one.
df_raw['Gender_Clean'] = df_raw['Gender'].map(gender_map)

# Step 5: Verify — check what we have now
print("After cleaning:")
print(df_raw['Gender_Clean'].value_counts())
print(f"\nAny NaN values? {df_raw['Gender_Clean'].isnull().sum()}")
```

```python
# Step 6: Drop the original dirty column now that we have a clean one
df_raw = df_raw.drop(columns=['Gender'])
df_raw = df_raw.rename(columns={'Gender_Clean': 'Gender'})

# Confirm the result
print("Column 'Gender' after cleaning:")
print(df_raw['Gender'].value_counts())
print(f"Data type: {df_raw['Gender'].dtype}")
df_raw.head()
```
